Real Estate Crime Statistics Data Preprocessing

```python
# Import Libraries 

import pandas as pd
import numpy as np
import os

print("Import Successful")
```

```python
# Find File Path

data_path = "Crime_Data"
print("Path Exists: ", os.path.exists(data_path))
```

```python
import os
os.getcwd()
```

```python
# Collect CSV files

csv_files = []
for root, dirs, files in os.walk(data_path):
    for file in files:
        if file.endswith('.csv'):
            csv_files.append(os.path.join(root, file))

print(f"Total CSV files found: {len(csv_files)}")
csv_files[:4] # Show four CSV files
```

```python
# Combine all CSV files together in one database

crime_list = [pd.read_csv(file) for file in csv_files]
crime_df = pd.concat(crime_list, ignore_index=True)

print("Files Combined")
crime_df.head()
```

```python
# Data Exploration

# View Data Information
crime_df.info()
crime_df.head()
```

```python
# Clean column Names
crime_df.columns = crime_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Removed unwanted columns
unwanted_cols = ['context', 'falls_within', 'lsoa_code']

cols_to_drop = [col for col in unwanted_cols if col in crime_df.columns]
print("Dropping columns:", cols_to_drop)

crime_df = crime_df.drop(columns=cols_to_drop, errors='ignore')

print("Columns removed")
print(crime_df.columns)
```

```python
# Check for Missing Data

print("Missing values per column: ")
print(crime_df.isnull().sum())

# Drop rows with missing coordinates (for mapping accuracy)
crime_df = crime_df.dropna(subset=['longitude', 'latitude'], how='any')

# Fill other missing text columns with 'Unknown'
crime_df = crime_df.fillna('Unknown')

print("Checked Missing Values")
```

```python
crime_df.columns
```

```python
# Convert Data types

# Convert 'month' column to datetime
crime_df['month'] = pd.to_datetime(crime_df['month'], errors='coerce')

# Extract year and month for analysis
crime_df['year'] = crime_df['month'].dt.year
crime_df['month_num'] = crime_df['month'].dt.month

print("Date Converted")

```

```python
# Checks 
print("Shape after cleaning:", crime_df.shape)
crime_df.info()
crime_df.head()
```

```python
# Save Cleaned Dataset

crime_df.to_csv('cleaned_crime_data.csv', index=False)
print("Cleaned dataset saved as 'cleaned_crime_data.csv'")
```

```python
# Sample Dataset

sample_df = crime_df.sample(n=min(5000, len(crime_df)), random_state=42)
sample_df.to_csv('sample_crime_data.csv', index=False)
print("Sample file 'sample_crime_data.csv' created for testing")
```
